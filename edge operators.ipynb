```python
import numpy as np
import matplotlib.pyplot as plt

from skimage.io import imread
from skimage.filters import roberts, sobel, scharr, prewitt
```

```python
img = imread("input/museu_do_amanha_01.jpg", as_gray=True)
op_roberts = roberts(img)
op_sobel = sobel(img)
```

```python
fig, ax = plt.subplots(ncols=3, sharex=True, sharey=True, figsize=(12, 8))

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title('Imagem original')

ax[1].imshow(op_roberts, cmap=plt.cm.gray)
ax[1].set_title('Operador de Roberts')

ax[2].imshow(op_sobel, cmap=plt.cm.gray)
ax[2].set_title('Operador de Sobel')

for a in ax:
    a.axis('off')

plt.tight_layout()
plt.show()
```

```python
img = imread("input/camera_01.jpg", as_gray=True)
op_roberts = roberts(img)
op_sobel = sobel(img)
op_scharr = scharr(img)
op_prewitt = prewitt(img)
```

```python
diff_scharr_prewitt = op_scharr - op_prewitt
diff_scharr_sobel = op_scharr - op_sobel
max_diff = np.max(np.maximum(diff_scharr_prewitt, diff_scharr_sobel))
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, sharex=True, sharey=True, figsize=(12, 12))

ax = axes.ravel()

ax[0].imshow(op_roberts, cmap=plt.cm.gray)
ax[0].set_title('Operador de Roberts')

ax[1].imshow(op_sobel, cmap=plt.cm.gray)
ax[1].set_title('Operador de Sobel')

ax[2].imshow(op_scharr, cmap=plt.cm.gray)
ax[2].set_title('Operador de Scharr')

ax[3].imshow(op_prewitt, cmap=plt.cm.gray)
ax[3].set_title('Operador de Prewitt')

ax[4].imshow(diff_scharr_prewitt, cmap=plt.cm.gray, vmax=max_diff)
ax[4].set_title('Diferença Scharr - Prewitt')

ax[5].imshow(diff_scharr_sobel, cmap=plt.cm.gray, vmax=max_diff)
ax[5].set_title('Diferença Scharr - Sobel')

for a in ax:
    a.axis('off')

plt.tight_layout()
plt.show()
```
